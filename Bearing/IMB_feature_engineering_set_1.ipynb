```python
# feature extraction
feature_matrix = np.zeros((1,9))
```

```python
def skew(x):
    
    n = len(x)
    third_moment = np.sum((x-np.mean(x))**3)/n
    s_3 = np.std(x, ddof = 1) ** 3
    return third_moment / s_3 
```

```python
def kurtosis(x):
    
    n = len(x)
    fourth_moment = np.sum((x-np.mean(x))**4) / n
    s_4 = np.std(x, ddof = 1) ** 4
    return fourth_moment / s_4 - 3
```

```python
feature_matrix[0,0] = np.max(temp)
feature_matrix[0,1] = np.min(temp)
feature_matrix[0,2] = np.mean(temp)
feature_matrix[0,3] = np.std(temp, ddof = 1)
feature_matrix[0,4] = np.sqrt(np.mean(temp ** 2))
feature_matrix[0,5] = skew(temp)
feature_matrix[0,6] = kurtosis(temp)
feature_matrix[0,7] = feature_matrix[0,0]/feature_matrix[0,4]
feature_matrix[0,8] = feature_matrix[0,4]/feature_matrix[0,2]
```

```python

time_features = ['Max','Min','Mean','Std','RMS','Skewness','Kurtosis','Crestfactor','Formfactor']
cols = ['1_a','1_b','2_a','2_b','3_a','3_b','4_a','4_b']
   
columns = [c+'_'+tf for c in cols for tf in time_features]
data = pd.DataFrame(columns=columns)
# data.index = [file_name[:-3]]
    
for file_name in os.listdir(file_path):
    
    # read dataset
    raw_data = pd.read_csv(os.path.join(file_path, file_name), sep='\t')      
    
    
    
    # time features
    Max = np.array(np.max(raw_data))
    Min = np.array(raw_data.min())
    Mean = np.array(raw_data.mean())
    Std = np.array(np.std(raw_data, ddof = 1))
    RMS = np.array(np.sqrt(np.mean(raw_data**2)))
    Skewness = np.array(skew(raw_data))
    Kurtosis = np.array(kurtosis(raw_data))
    Crestfactor = np.array(Max/RMS)
    Formfactor = np.array(RMS/Mean)
        
    
#     # time features
#     feature_matrix[0,0] = np.max(temp)
#     feature_matrix[0,1] = np.min(temp)
#     feature_matrix[0,2] = np.mean(temp)
#     feature_matrix[0,3] = np.std(temp, ddof = 1)
#     feature_matrix[0,4] = np.sqrt(np.mean(temp ** 2))
#     feature_matrix[0,5] = skew(temp)
#     feature_matrix[0,6] = kurtosis(temp)
#     feature_matrix[0,7] = feature_matrix[0,0]/feature_matrix[0,4]
#     feature_matrix[0,8] = feature_matrix[0,4]/feature_matrix[0,2]
        
        
    Max = pd.DataFrame(Max.reshape(1,8), columns=[c+'_Max' for c in cols])
    Min = pd.DataFrame(Min.reshape(1,8), columns=[c+'_Min' for c in cols])
    Mean = pd.DataFrame(Mean.reshape(1,8), columns=[c+'_Mean' for c in cols])
    Std = pd.DataFrame(Std.reshape(1,8), columns=[c+'_Std' for c in cols])
    RMS = pd.DataFrame(RMS.reshape(1,8), columns=[c+'_RMS' for c in cols])
    Skewness = pd.DataFrame(Skewness.reshape(1,8), columns=[c+'_Skewness' for c in cols])
    Kurtosis = pd.DataFrame(Kurtosis.reshape(1,8), columns=[c+'_Kurtosis' for c in cols])
    Crestfactor = pd.DataFrame(Crestfactor.reshape(1,8), columns=[c+'_Crestfactor' for c in cols])
    Formfactor = pd.DataFrame(Formfactor.reshape(1,8), columns=[c+'_Formfactor' for c in cols])
    
    
    Max.index = [file_name]
    Min.index = [file_name]
    Mean.index = [file_name]
    Std.index = [file_name]
    RMS.index = [file_name]
    Skewness.index = [file_name]
    Kurtosis.index = [file_name]
    Crestfactor.index = [file_name]
    Formfactor.index = [file_name]
    
    merge = pd.concat([Max,Min,Mean,Std,RMS,Skewness,Kurtosis,Crestfactor,Formfactor], axis = 1)
    
    data=data.append(merge)
    
    
```
